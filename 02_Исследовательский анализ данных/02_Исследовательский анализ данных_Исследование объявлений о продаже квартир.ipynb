## Предобработка данных

Выведем на экран названия столбцов:

```python
data.columns
```

Переименуем следующие столбцы:

```python
data = data.rename(
    columns={
        'cityCenters_nearest': 'city_centers_nearest',
        'parks_around3000': 'parks_around_3000',
        'ponds_around3000': 'ponds_around_3000',
    }
)
```

Проверим результат:

```python
data.columns
```

Посмотим на уникальные названия населенных пунктов:

```python
data['locality_name'].sort_values().unique()
```

Сделаем несколько замен:

```python
# заменим все ё на е:
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')

# заменим "городской поселок" на "поселок городского типа":
data['locality_name'] = data['locality_name'].str.replace('городской поселок', 'поселок городского типа')

# заменим "садоводческое некоммерческое товарищество" на "садовое товарищество"
data['locality_name'] = data['locality_name'].str.replace('садоводческое некоммерческое товарищество', 'садовое товарищество')

# заменим "поселок при железнодорожной станции" на "поселок станции"
data['locality_name'] = data['locality_name'].str.replace('поселок при железнодорожной станции', 'поселок станции')
data['locality_name'] = data['locality_name'].str.replace('поселок при станции', 'поселок станции')

# заменим "поселок Мурино" на "Мурино"
data['locality_name'] = data['locality_name'].str.replace('поселок Мурино', 'Мурино')

# заменим "деревня Кудрово" на "Кудрово"
data['locality_name'] = data['locality_name'].str.replace('деревня Кудрово', 'Кудрово')

# выведем результат
data['locality_name'].sort_values().unique()
```

Выведем общее число пропущенных значений при помощи двух методов:

```python
data.isna().sum()
```

Удалим строки без общего количества этажей `floors_total` и названия местности `locality_name`, т.к. количество пропусков в них несущественное, а без этих данных невозможно будет рассматривать квартиры в дальнейшем:

```python
data.dropna(subset=['floors_total'], inplace=True)
data.dropna(subset=['locality_name'], inplace=True)
```

В колонке с количеством балконов заменим NaN на 0 и приведем к целому значению:

```python
data['balcony'] = data['balcony'].fillna(value=0).astype(int)
```

Пропущенные значения в столбце `days_exposition` могут означать, что объявление не закрыто на момент выгрузки данных - не будем менять данные.

Данные в `parks_around_3000` и `ponds_around_3000` должны заполняться автоматически, но по какой-то причине этого не произошло - для целей проекта мы можем оставить их как есть или заменить на "unknown":

```python
columns_to_change = ['parks_around_3000', 'ponds_around_3000']

for column in columns_to_change:
    data[column] = data[column].fillna('unknown')
display(data[columns_to_change].head(20))
```

Также зменим общее число этажей и стоимость квартиры на целое значение. В стоимости нас интересует порядок цифр, точность до копеек нам не нужна.

```python
data['floors_total'] = data['floors_total'].astype(int)
data['last_price'] = data['last_price'].astype(int)
```

Дату публикации приводим к формату даты (сейчас тип данных object):

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Узнаем количество значений по апартаментам:

```python
data['is_apartment'].value_counts()
```

То есть в списке из заполненных значений всего 50 апартаментов. При этом всего пропущенных значений для этого параметра 20924 шт. Можем предположить, что пропуски означают, что параметр неприменим, и заменить NaN на `False`. Тогдп логика типа данных поменяется на bool:

```python
data['is_apartment'] = data['is_apartment'].fillna(value=False)
```

Проверим типы данных еще раз:

```python
data.info()
```

Таким образом, типы данных теперь отображаются корректно (не принимаем во внимание `parks_around_3000` и `ponds_around_3000`). Продолжим убирать пропуски - проверим, в каких данных они остались:

```python
data.isna().sum()
```

У нас осталось осталось два типа данных с пропусками:
- параметры квартиры, по которым пропуски можно заполнить медианными значениями (площадь помещений и высота потолков)
- те, которые были получены автоматически на основе картографических данных

По автоматически полученным данным стоило бы установить причину пропусков и восстановить их, обратившись к отправителю данных. Для целей проекта, так как этих пропусков достаточно много, мы не можем их удалить, поэтому заменим на условное 999999:

```python
# создадим список distances_to_replace с названиями столбцов, в которых будем заменять пропущенные звачения на "заглушку":
columns_to_replace = ['airports_nearest', 'city_centers_nearest', 'parks_nearest', 'ponds_nearest']

# переберем названия столбцов в цикле и заменим пропущенные значения на 999999:
for column in columns_to_replace:
    data[column] = data[column].fillna(value=999999)
display(data[columns_to_replace].head(20))
```

Посмотрим даные столбца `ceiling_height`. Для начала выведем перечень его уникальных значений в порядке возрастания:

```python
data['ceiling_height'].sort_values().unique()
```

Сделаем следующие допущения:
- Возьмем за условный минимум высоту потолков 2.4 м ("хрущевки").
- В некоторых значениях похоже некорректно указана размерность (24, 25, 26, 27, 27.5 и 32 м) - разделим значения более 24 на 10.
- Допустим, что высота потолков не может быть двухзначной величиной.

Исходя из этих параметров заменим значения высоты потолков в таблице:

```python
# Разделим высоты более 24 метров на 10:
data.loc[data['ceiling_height'] >= 24, 'ceiling_height'] = data['ceiling_height']/10

# уберем значение 100 как ошибочнoе:
data = data[data['ceiling_height'] != 100]
```

Проверим, сколько значений каждой высоты осталось в начале и конце списка:

```python
pivot_height = data.pivot_table(index=['ceiling_height'], columns='is_apartment', values='floor', aggfunc='count')
display(pivot_height.head(10))
display(pivot_height.tail(10))
```

```python
# уберем строки со значением высот менее 2.4 и более 6 метров как ошибочные
# квартиры с высокими потолками могут быть двухуровневыми, но их в любом случае немного,
# поэтому примем за условный максимум 6 м
# через data.loc убираются Nan, поэтому удаляем значения вручную

data = data[
    (data['ceiling_height'] != 1) &
    (data['ceiling_height'] != 1.2) &
    (data['ceiling_height'] != 1.75) &
    (data['ceiling_height'] != 2) &
    (data['ceiling_height'] != 2.2) &
    (data['ceiling_height'] != 2.25) &
    (data['ceiling_height'] != 2.3) &
    (data['ceiling_height'] != 2.34) &
    (data['ceiling_height'] != 8) &
    (data['ceiling_height'] != 8.3) &
    (data['ceiling_height'] != 10) &
    (data['ceiling_height'] != 10.3) &
    (data['ceiling_height'] != 14) &
    (data['ceiling_height'] != 20) &
    (data['ceiling_height'] != 22.6)
]

#data['ceiling_height'].sort_values().unique()
```

### Средняя цена одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений 

Посчитаем среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений. Выделим населённые пункты с самой высокой и низкой стоимостью квадратного метра. Воспользуемся данными столбца `locality_name`.

```python
# определим 10 населенных пунктов с наибольшим числом объявлений
top_10 = data['locality_name'].value_counts().head(10)

# создадим сводную таблицу, используя фильтр - так, чтобы местоположение входило в список top_10
(
    data.query('locality_name in @top_10.index').pivot_table(index='locality_name', values='meter_price')
    # приведем к целочисленному типу
    .astype(int)
    # отсортируем по убыванию стоимости за квадратный метр
    .sort_values('meter_price', ascending=False)
)
```

**Вывод:**

Как и ожидалось, наибольшая цена за квадратный метр в Санкт-Петербурге и Пушкине. Кудрово, Мурино и Шушары очень популярны для жизни, поэтому они также попали в список. Самая низкая цена за квадратный метр в Выборге - он дальше всего находится от Санкт-Петербурга.

### Как стоимость объектов зависит от расстояния до центра города

Для начала определим, что будем считать центром - выведем график со средней ценой для каждого километра удаленности от центра:

```python
(
    data.query('locality_name == "Санкт-Петербург"')
    .pivot_table(index='city_center_km', values='last_price')
    .plot(grid=True, style='o-', xlim=(0,25), figsize=(10, 5))
)
plt.show()
```

Резкое падение цены наблюдается примерно после 7-го километра от центра. Определим эту зону как центральную, и будем рассматривать ее далее.

Выделим квартиры в Санкт-Петербурге с помощью столбца `locality_name` и вычислим среднюю цену каждого километра. Посмотрим, как стоимость объектов зависит от расстояния до центра города.

```python
# создадим список квартир, расположенных в Санкт-Петербурге, и рассмотрим те, что удалены от центра до 7 км
locality_spb = data.query('(locality_name == "Санкт-Петербург") & (city_center_km <= 7)')

# создадим сводную таблицу со стоимостью объекта и удалённостью до центра
# будем рассматривать не укрупненный показатель city_center_km, а более детальный city_centers_nearest
(
    locality_spb.pivot_table(index='city_centers_nearest', values='last_price', aggfunc='mean')
    # построим диаграмму рассеивания
    .plot(grid=True, style='o', figsize=(5, 5))
)
plt.show()
print('Корреляция стоимости объекта и расстояния до центра:', locality_spb['city_center_km'].corr(locality_spb['last_price']))
```

**Вывод:**

Зависимости стоимости оъекта и расстояния до центра практически нет, но слабая тенденция такова, что чем дальше от центра, тем дешевле жилье (об этом говорит отрицательное значение коэффициента корреляции).
