**Описание проекта**

Bнтернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

У нас есть данные до 2016 года. Представим, что сейчас декабрь 2016 г., и нам надо спланировать кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируем ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

### Сколько игр выпускалось в разные годы

Посмотрим, сколько игр выпускалось в разные годы, и важны ли данные за все периоды.

```python
(
    data.pivot_table(index='year_of_release', values='name', aggfunc='count')
    .plot(
        title='Количество выпущенных игр по годам',
        xlabel='Год выпуска',
        ylabel='Количество игр',
        grid=True,
        style='o-',
        figsize=(10, 5))
    
)
plt.show()
```

**Вывод:**

С 1980 до 1994 года выпускалось лишь небольшое число игр в год. Их количество начало постепенно увеличиваться к 2000 году, а затем произошел резкий подъем индустрии, который к 2008-2009 годам достиг своего максимума. После чего последовал такой же резкий спад, и так до 2013 года. Далее возможны различные варианты развития сбытий, т.к. были как рост количества выпущенных игр в 2013-2015гг., так и последующий спад к 2016 году.

Возможно, резкое падение количества новых игр связано с тем, что пользователи стали активнее использовать мобильные приложения, таким образом рынок переориентировался на новые носители.

Ситуация неоднозначна, но в целом для анализа интересен период с 2012 года, когда после резкого спада показатели индустрии начали стабилизироваться. Основной фактор здесь — эти данные помогут построить прогноз на 2017 год.

### Как менялись продажи по платформам

Посмотрим, как менялись продажи по платформам за весь период. Для начала выберем платформы с наибольшими суммарными продажами за весь период.

```python
(
    data.pivot_table(index='platform', values='total_sales', aggfunc='sum')
    # отсортируем по убыванию продаж
    .sort_values('total_sales', ascending=False)
    # построим столбчатую диаграмму
    .plot(
        title='Рейтинг продаж',
        xlabel='Платформа',
        ylabel='Общие продажи, млн.проданных копий',
        kind='bar', rot=0, grid=True, figsize=(15, 5))
)
plt.show()
```

Платформы с наибольшими суммарными продажами за весь период - это PS2, Х360, PS3, Wii, DS и PS, которые расшифровываются как
- PS2 - PlayStation 2;
- Х360 - Xbox 360;
- PS3 - PlayStation 3;
- Wii - Nintendo Wii;
- DS - Nintendo DS;
- PS - PlayStation.

```python
# выведем показатели продаж топ-6 платформ в таблице
(
    data.pivot_table(index='platform', values='total_sales', aggfunc='sum')
    .sort_values('total_sales', ascending=False)
    .head(6)
)
```

Построим по ним распределение по годам. Также посмотрим, за какой характерный срок появляются новые и исчезают старые платформы.

```python
(
    data.query('platform == ["PS2", "X360", "PS3", "Wii", "DS", "PS"]')
    .pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')
    .plot(
        title='Рейтинг продаж по топовым платформам',
        xlabel='Год выпуска',
        ylabel='Общие продажи, млн.проданных копий',
        xlim=(1993, 2017), grid=True, style='o-', figsize=(20, 10))
)
plt.show()
```

**Вывод:**

Платформы набирали популярность примерно за 4 года. Срок жизни платформы около 10 лет. Не все платформы из самых прибыльных за весь период продолжили существование в 2016 году, поэтому имеет смысл определить "актуальный период" для дальнейшего анализа.

### Определим "актуальный период"

Кроме того, предыдущий анализ показал, что платформы достигают своего пика популярности где-то на 4-й год. Таким образом, после 2012 года (2016 год минус 4) какие-то платформы будут находиться на пике популярности, либо будут зарождаться.

### Выберем несколько потенциально прибыльных платформ

```python
(
    data.pivot_table(index='platform', values='total_sales', aggfunc='sum')
    # отсортируем по убыванию продаж
    .sort_values('total_sales', ascending=False)
    # построим столбчатую диаграмму
    .plot(
        title='Обновленный рейтинг продаж',
        xlabel='Платформа',
        ylabel='Общие продажи, млн.проданных копий',
        kind='bar', rot=0, grid=True, figsize=(15, 5))
)
plt.show()
```

Платформы с наибольшими суммарными продажами теперь - PS4, PS3, Х360, 3DS и XOne, то есть:
- PS4 - PlayStation 4;
- PS3 - PlayStation 3;
- Х360 - Xbox 360;
- 3DS - Nintendo 3DS;
- XOne - Xbox One.

Построим по ним распределение по годам:

### Как влияют на продажи отзывы пользователей и критиков

Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Возьмем платформы из топ-5 (с наибольшими продажами), построим диаграммы рассеяния и посчитаем корреляцию между отзывами и продажами. Сопоставим влияние отзывов критиков и пользователей:

```python
# напишем функцию, которая будет строить графики рассеивания и считать корреляцию
def corr_platform (platform_name):
    platform = data[data['platform'] == platform_name]
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
    fig.suptitle(platform_name, fontsize=15)
    
    # данные для графика
    sns.scatterplot(x='user_score', y='total_sales', data=platform, ax=ax[0])
    sns.scatterplot(x='critic_score', y='total_sales', data=platform, ax=ax[1])
    
    # названия таблицы и осей
    ax[0].set(title='Влияние оценки пользователей',
              xlabel='Оценка пользователей',
              ylabel='Общие продажи')
    ax[1].set(title='Влияние оценки критиков',
              xlabel='Оценка критиков',
              ylabel='Общие продажи')
    plt.show()
    
    # посчитаем корреляцию
    user_corr = platform['user_score'].corr(platform['total_sales'])
    critic_corr = platform['critic_score'].corr(platform['total_sales'])
    
    # выведем ее значения на экран
    print('Корреляция между отзывами пользователей и продажами в целом:', user_corr)
    print('Корреляция между отзывами критиков и продажами в целом:', critic_corr)
```

```python
# при помощи цикла выведем на экран все графики для топ-5 платформ
for platform in ["PS4", "PS3", "X360", "3DS", "XOne"]:
    corr_platform (platform)
```

**Вывод:**

Получается, что при оценках критиков от 0 до 70 (из 100 возможных) это практически не влияет на продажи. Слабая зависимость была, когда критики ставили игре оценку выше 70 баллов. Эти выводы характерны для продажи на всех платформах из топ-5.

Зависимость продаж игр от оценки пользователей отсутствует. При оценке выше 5 (из 10 возможных) уровень продаж распределялся примерно одинаково. Эти выводы характерны для продажи и на всех платформах из топ-5.

Но в целом можно сказать, что оценка критиков играет для покупателя бОльшую роль.

## Проверка гипотез

Проверим следующие гипотезы:
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Для этого:
- зададим пороговое значение alpha;
- сформулируем нулевую (Н0) и альтернативную (Н1) гипотезы;
- выберем и обоснуем критерий для проверки гипотезы.

### Гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Уровень значимости для alpha установим 0,05. Мы будем сравивать с ним значение вероятности для нулевой гипотезы (p-value). Если p-value окажется меньше alpha, полученный результат будет являться статистически значимым, и мы примем нулевую гипотезу.

Сформулируем нулевую и анльтернативную гипотезы:

- Н0: Средний пользовательский рейтинг платформ Xbox One и PC одинаковые.
- H1: Средний пользовательский рейтинг платформ Xbox One и PC различаются.

Так как выборки независимы между собой, будем использовать t-тест (t-распределение Стьюдента).

```python
# отфильтруем платформы XOne и PC при пользовательском рейтинге больше 0
data_xone = data.query('platform == "XOne" and user_score > 0')
data_pc = data.query('platform == "PC" and user_score > 0')

# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    data_xone['user_score'],
    data_pc['user_score'],
    equal_var=False)
    

# alpha = значение уровня значимости
alpha = .05

# вывод значения p-value на экран
print('p-value:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
    
print('Средний пользовательские рейтинг платформы XOne:', data_xone['user_score'].mean())
print('Средний пользовательские рейтинг платформы PC:', data_pc['user_score'].mean())
```

**Вывод:**

Значение p-value равно 55%. Это означает, что с вероятностью 55% рейтинги двух платформ совпадают. Таким образом, не получилось отвергнуть нулевую гипотезу - средние пользовательские рейтинги платформ Xbox One и PC действительно одинаковые. Это подтверждают и сами расчеты средних рейтингов.

### Гипотеза: средние пользовательские рейтинги жанров Action и Sports разные

Сформулируем нулевую и анльтернативную гипотезы:

- Н0: Средние пользовательские рейтинги жанров Action и Sports одинаковые.
- H1: Средние пользовательские рейтинги жанров Action и Sports разные.

Так как выборки независимы между собой, будем использовать t-тест (t-распределение Стьюдента).

```python
# отфильтруем жанры Action и Sports при пользовательском рейтинге больше 0
data_action = data.query('genre == "Action" and user_score > 0')
data_sports = data.query('genre == "Sports" and user_score > 0')

# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    data_action['user_score'],
    data_sports['user_score'],
    equal_var=False)
    

# alpha = значение уровня значимости
alpha = .05

# вывод значения p-value на экран
print('p-value:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
    
print('Средний пользовательский рейтинг жанра Action:', data_action['user_score'].mean())
print('Средний пользовательский рейтинг жанра Sports:', data_sports['user_score'].mean())
```

**Вывод:**

В результате расчетов мы отвергли нулевую гипотезу, то есть средние пользовательские рейтинги жанров Action и Sports на самом деле разные. Это подтверждают и сами расчеты средних рейтингов.

- Европа: предпочитают игры на платформе PlayStation 4 в жанре "Action" с рейтингом "Для взрослых". Можно обратить внимание на PlayStation 3 и Xbox 360, а также на жанры "Shooter" и "Sports", т.к. они также пользуются популярностью.
- Япония: пользователи предпочитают ролевые игры на платформе Nintendo 3DS, можно также обратить внимание на платформу PlayStation 3 и жанр "Action". 

Продажи не зависят от отзывов пользователей, а критики оказывают слабое влияние. Можно принимать во внимание оценку последних, если она была выше 70 баллов (из 100 возможных).
